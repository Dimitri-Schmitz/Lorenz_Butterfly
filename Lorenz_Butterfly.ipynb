```python
#This program reads in an electronic signal generated from a set of 'chaotic'
  #sinusodual equations, so sensitive to initial conditions that repeating
  #the same procedure produced a different result. When viewed as a function
  #of time, the signals were chaotic and unrepeatable. After undergoing a Fast
  #Fourier Transform to become a function of frequency instead of time, the
  #data takes on the characteristic Lorenz Butterfly shape.

#imports
import matplotlib.pyplot as plt
import numpy as np
#read in data from local file (available at https://www.dropbox.com/scl/fi/w0x8xgpki63qjxw5fwr7l/lorenz_data.txt?rlkey=b5769c4xlqb9mevnlibe90l03&st=behkl48e&dl=0)
xdata = np.genfromtxt('lorenz_data.txt', usecols=[0], delimiter = ',', unpack=True)
ydata = np.genfromtxt('lorenz_data.txt', usecols=[1], delimiter = ',', unpack=True)
#plot data
fig, ax = plt.subplots()
ax.plot(xdata, ydata, linewidth=0.02)
plt.xlabel('V_x (V)')
plt.ylabel('V_z (V)')
plt.rc('font', size=20)
```

```python
#To see just the butterlfy, plot as a scatter plot
fig = plt.scatter(xdata, ydata)
plt.xlabel('V_x (V)')
plt.ylabel('V_z (V)')
plt.rc('font', size=5)
```

```python
#To see the butterfly build over time, plot the data four times, adding more
    #during each iteration. Start with finding the number of rows in the data:
count=0
file=open('lorenz_data.txt')
data=file.read()
point = data.split(",")
 
for i in point:
    if i:
        count += 1
print("This is the number of lines in the file:", count)
print("This is 1/4th of the rows:", count/4)
```

```python
#Make three data sets, adding 1/4th of the data each time
xdata1 = np.genfromtxt('lorenz_data.txt', usecols=[0], delimiter = ',', unpack=True, max_rows=625)
ydata1 = np.genfromtxt('lorenz_data.txt', usecols=[1], delimiter = ',', unpack=True, max_rows=625)
xdata2 = np.genfromtxt('lorenz_data.txt', usecols=[0], delimiter = ',', unpack=True, max_rows=1250)
ydata2 = np.genfromtxt('lorenz_data.txt', usecols=[1], delimiter = ',', unpack=True, max_rows=1250)
xdata3 = np.genfromtxt('lorenz_data.txt', usecols=[0], delimiter = ',', unpack=True, max_rows=1875)
ydata3 = np.genfromtxt('lorenz_data.txt', usecols=[1], delimiter = ',', unpack=True, max_rows=1875)
#Create a 2x2 plot grid and plot the data
fig,axs =plt.subplots(2,2)
axs[0, 0].plot(xdata1, ydata1, linewidth=0.02)
axs[0, 0].set_title('1/4th Data')
axs[0, 1].plot(xdata2, ydata2, linewidth=0.02)
axs[0, 1].set_title('1/2th Data')
axs[1,0].plot(xdata3, ydata3, linewidth=0.02)
axs[1,0].set_title('3/4th Data')
axs[1,1].plot(xdata, ydata, linewidth=0.02)
axs[1,1].set_title('Full Data')
plt.rcParams.update({'font.size':10})
for ax in axs.flat:
    ax.set(xlabel='V_x (V)', ylabel='V_z (V)')
    ax.label_outer()
```
